```julia
include("MyLib.jl")
using Plots
using .MyLib
function func(x)
    return x^2
end
function CheckMydiff()
    x=0:0.1:2
    y=func.(x)
    h=x[2]-x[1]
    dydx=MyLib.diff1st(h,y)
    ddyddx=MyLib.diff2nd(h,y)
    plot(xlabel="x",ylabel="y",title="check My diff",legend=:bottomright)
    plot!(x,y,label="y=x")
    plot!(x,dydx,label="y=dydx")
    plot!(x,ddyddx)
end
CheckMydiff()
```
